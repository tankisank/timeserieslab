# Free hand method

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data
x = np.array([2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019])
y = np.array([150, 145, 155, 160, 162, 165, 172, 178, 180])

# Plot the data
plt.plot(x, y, 'o')

# Fit a freehand curve
curve_fit = np.polyfit(x, y, 3)
p = np.poly1d(curve_fit)
x_new = np.linspace(x[0], x[-1], 50)
y_new = p(x_new)
plt.plot(x_new, y_new, '-')

# Find the trend line
mid = (x[0] + x[1]) / 2
trend_slope = (y[1] - y[0]) / (x[1] - x[0])
trend_intercept = y[0] - trend_slope * mid
trend_line = trend_slope * x + trend_intercept
plt.plot(x, trend_line, '--')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales vs Year')

# Show the plot
plt.show()
```

# Semi-Average(Even)

```python
import matplotlib.pyplot as plt

# Given data
year = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
sales = [412, 438, 444, 470, 482, 490, 500, 510, 512, 520]

# Compute the semi-average and mid-year values
semi_avg = []
mid_year = []
for i in range(0, len(sales), 2):
    tot = sales[i] + sales[i+1]
    avg = tot / 2
    semi_avg.append(avg)
    mid = (year[i+1] + year[i]) / 2
    mid_year.append(mid)

# Plot the original data and trendline
plt.plot(year, sales, 'o', label='Original data')
plt.plot(mid_year, semi_avg, 'r-', label='Trendline')

# Set axis labels and title
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Semi-Average Trendline')

# Display legend and plot
plt.legend()
plt.show()
```

# Semi-Avg(Odd)

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the data
year = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
sales = [412, 438, 444, 470, 482, 490, 500, 510, 512]

# Step 2: Plot the data
plt.plot(year, sales, 'o', label='Sales Data')

# Step 3: Draw the trend line
trend_line = np.polyfit(year, sales, 1)
trend_line_fn = np.poly1d(trend_line)
mid_year = (year[0] + year[1]) / 2
trend_start = trend_line_fn(mid_year)
trend_end = trend_line_fn(mid_year + len(sales) - 1)
plt.plot([mid_year, mid_year + len(sales) - 1], [trend_start, trend_end], label='Trend Line')

# Step 4: Draw the table
print("Year\tSales\tSimi_Tot\tSemi_Av\tMid_Year")
print("------------------------------------------------")
for i in range(len(year)):
    if i == 0:
        simi_tot = sales[i]
        semi_av = simi_tot / 1
    elif i == len(year) - 1:
        simi_tot = sales[i-1] + sales[i]
        semi_av = simi_tot / 2
    else:
        simi_tot = sales[i-1] + sales[i]
        semi_av = simi_tot / 2
    if i % 2 == 0:
        mid_year = year[int(i/2) + 1]
    else:
        mid_year = (year[i] + year[i+1]) / 2
    print(f"{year[i]}\t{sales[i]}\t{simi_tot}\t{semi_av:.2f}\t{mid_year:.1f}")

# Step 7: Plot the averages values with Mid_year
avg_sales = []
avg_year = []
for i in range(len(year)):
    if i == 0:
        simi_tot = sales[i]
        semi_av = simi_tot / 1
    elif i == len(year) - 1:
        simi_tot = sales[i-1] + sales[i]
        semi_av = simi_tot / 2
    else:
        simi_tot = sales[i-1] + sales[i]
        semi_av = simi_tot / 2
    if i % 2 == 0:
        mid_year = year[int(i/2) + 1]
    else:
        mid_year = (year[i] + year[i+1]) / 2
    avg_sales.append(semi_av)
    avg_year.append(mid_year)
plt.plot(avg_year, avg_sales, 'x', label='Semi-Average Values')

plt.legend()
plt.show()
```

# MOving Average(Odd)


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the data
year = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
cyclic_fluctuations = [2, 1, 0, -2, -1, 0, 2, 1, 0, -1]

# Step 2: Define the moving window size
moving_window_3 = 3
moving_window_5 = 5

# Step 3: Compute the moving averages using the odd method
moving_avg_3 = np.convolve(cyclic_fluctuations, np.ones(moving_window_3), 'valid') / moving_window_3
moving_avg_5 = np.convolve(cyclic_fluctuations, np.ones(moving_window_5), 'valid') / moving_window_5

# Step 4: Define the table headers
print("{:<6} {:<8} {:<8} {:<8}".format("Year", "Profit", "Grp_Tot.", "Grp_Avg"))

# Step 5: Compute the group totals and group averages for 3 yearly moving average
for i in range(len(moving_avg_3)):
    if i == 0:
        group_total = sum(cyclic_fluctuations[:3])
    else:
        group_total = group_total - cyclic_fluctuations[i-1] + cyclic_fluctuations[i+2]
    
    group_average = group_total / moving_window_3
    
    # Step 6: Compute the mid-year and print the results
    mid_year = year[i+1]
    print("{:<6} {:<8} {:<8.1f} {:<8.1f}".format(mid_year, moving_avg_3[i], group_total, group_average))

# Step 7: Compute the group totals and group averages for 5 yearly moving average
for i in range(len(moving_avg_5)):
    if i == 0:
        group_total = sum(cyclic_fluctuations[:5])
    else:
        group_total = group_total - cyclic_fluctuations[i-1] + cyclic_fluctuations[i+4]
    
    group_average = group_total / moving_window_5
    
    # Step 8: Compute the mid-year and plot the results
    mid_year = year[i+2]
    plt.scatter(mid_year, group_average, color='red')
    print("{:<6} {:<8} {:<8.1f} {:<8.1f}".format(mid_year, moving_avg_5[i], group_total, group_average))

# Plot the data and moving averages
plt.plot(year, cyclic_fluctuations, label='Cyclic Fluctuations')
plt.plot(year[moving_window_3-1:], moving_avg_3, label='3 Yearly Moving Avg.')
plt.plot(year[moving_window_5-1:], moving_avg_5, label='5 Yearly Moving Avg.')
plt.legend()
plt.show()
```

# Moving Avg(Even)

```python
import matplotlib.pyplot as plt

# data
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
fluctuations = [7, 8, 10, 9, 11, 10, 8, 9, 7, 8, 6]

# calculate 4-year moving average
ma_4yr = []
for i in range(len(fluctuations) - 3):
    ma_4yr.append(sum(fluctuations[i:i+4]) / 4)
print("4 moving average:\n",ma_4yr)
# plot the graph
plt.plot(years, fluctuations, label="Cyclic Fluctuations")
plt.plot(years[2:-1], ma_4yr, label="4-year Moving Average")
plt.title("Cyclic Fluctuations and 4-year Moving Average")
plt.xlabel("Year")
plt.ylabel("Fluctuations")
plt.legend()
plt.show()

```

# Weighted Moving Avg(Odd)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data and the weights
year = np.array([2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018])
values = np.array([2, 1, 6, 5, 3, 7, 2, 1])
weights = np.array([1, 4, 1])

# Compute the weighted moving averages
n = len(values)
yry_ma = np.zeros(n)
for i in range(1, n-1):
    yry_ma[i] = (weights[0]*values[i-1] + weights[1]*values[i] + weights[2]*values[i+1]) / np.sum(weights)

# Print the results
print("Year\tValues\t3 Yry MA")
for i in range(n):
    print(year[i], "\t", values[i], "\t", yry_ma[i])

# Plot the results
plt.plot(year, values, label="Values")
plt.plot(year[1:-1], yry_ma[1:-1], label="3 Yry MA")
plt.legend()
plt.show()
```

# Calculate the n-period


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data
year = np.arange(2011, 2029)
sales = np.array([41, 43, 51, 47, 53, 66, 30, 51, 77, 30, 43, 47, 17, 28, 50, 40, 15, 22])

# Compute the 3-year moving average
n = 3
weights = np.ones(n) / n
ma = np.convolve(sales, weights, mode='valid')

# Plot the data and trendline
plt.plot(year, sales, 'o-', label='Sales')
plt.plot(year[n-1:], ma, 'r--', label='Moving Average')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

# Least sq method - Direct method

```python
import numpy as np
import matplotlib.pyplot as plt

# Input data
year = np.array([2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018])
values = np.array([5, 7, 12, 10, 13, 17, 12, 11])

# Compute the coefficients of the linear regression line
coefficients = np.polyfit(year, values, 1)
slope, intercept = coefficients

# Compute the trendline
trendline = slope * year + intercept

# Plot the data and trendline
plt.plot(year, values, 'o', label='Data')
plt.plot(year, trendline, label='Trendline')
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('Least Squares Trendline')
plt.legend()
plt.show()
```

# Least sq - Indirect method

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Input data
year = np.array([2011, 2012, 2013, 2014, 2015, 2016, 2017])
values = np.array([80, 90, 92, 83, 94, 99, 92])

# Compute the trendline using linear regression
slope, intercept, r_value, p_value, std_err = linregress(year, values)
trendline = intercept + slope * year

# Plot the data and trendline
plt.plot(year, values, 'o', label='Data')
plt.plot(year, trendline, label='Trendline')
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('Least Squares Trendline')
plt.legend()
plt.show()

# Compute the trend value for 2011
trend_value = intercept + slope * 2011
print('Trend value for 2011:', trend_value)

```

# Lags and Autocorrelation

```python
import pandas as pd

original_data = [2.2, 7.16, 9.8, 11.93, 12.41, 12.51, 12.63, 14.33, 15, 15.54, 16.1, 17.91, 18, 18.28, 19.68, 20.04, 20.73, 21, 22.11, 24, 27, 30, 33, 36]

df = pd.DataFrame({'Original': original_data,
                   'lag1': ['NaN'] + original_data[:-1],
                   'lag2': ['NaN', 'NaN'] + original_data[:-2],
                   'lag3': ['NaN', 'NaN', 'NaN'] + original_data[:-3]})
df
```

```python
import pandas as pd

# create the dataframe with lagged values
original_data = [2.2,7.16,9.8,11.93,12.41,12.51,12.63,14.33,15,15.54,16.1,17.91,18,18.28,19.68,20.04,20.73,21,22.11,24,27,30,33,36]

df = pd.DataFrame(original_data, columns=['original'])
for i in range(1, 4):
    df['lag_{}'.format(i)] = df['original'].shift(i)

# calculate autocorrelation
autocorr = df.corr()
print(autocorr)
```

# Seasonality Index


```python
import numpy as np

# Input data
Year = [2003, 2004, 2005]
Q1 = [72, 75, 76]
Q2 = [64, 66, 68]
Q3 = [63, 64, 67]
Q4 = [75, 89, 95]

# Compute the mean of each quarter
mean_Q1 = np.mean(Q1)
mean_Q2 = np.mean(Q2)
mean_Q3 = np.mean(Q3)
mean_Q4 = np.mean(Q4)

# Compute the seasonal index for each quarter
index_Q1 = mean_Q1 / np.mean([mean_Q1, mean_Q2, mean_Q3, mean_Q4])
index_Q2 = mean_Q2 / np.mean([mean_Q1, mean_Q2, mean_Q3, mean_Q4])
index_Q3 = mean_Q3 / np.mean([mean_Q1, mean_Q2, mean_Q3, mean_Q4])
index_Q4 = mean_Q4 / np.mean([mean_Q1, mean_Q2, mean_Q3, mean_Q4])

# Print the seasonal index for each quarter
print("Seasonal Index for Q1:", round(index_Q1, 2))
print("Seasonal Index for Q2:", round(index_Q2, 2))
print("Seasonal Index for Q3:", round(index_Q3, 2))
print("Seasonal Index for Q4:", round(index_Q4, 2))
```

# Ratio- to - Seasonality Index'

```python
Year = [2003, 2004, 2005]
Q1 = [72, 75, 76]
Q2 = [64, 66, 68]
Q3 = [63, 64, 67]
Q4 = [75, 89, 95]

# Calculate the average sales for each quarter
Q1_avg = sum(Q1) / len(Q1)
Q2_avg = sum(Q2) / len(Q2)
Q3_avg = sum(Q3) / len(Q3)
Q4_avg = sum(Q4) / len(Q4)

# Calculate the overall average sales for the year
overall_avg = sum(Q1+Q2+Q3+Q4) / len(Year)

# Calculate the ratio-to-seasonality index for each quarter
Q1_ratio = Q1_avg / overall_avg
Q2_ratio = Q2_avg / overall_avg
Q3_ratio = Q3_avg / overall_avg
Q4_ratio = Q4_avg / overall_avg

print("Ratio-to-seasonality index for Q1: {:.2f}".format(Q1_ratio))
print("Ratio-to-seasonality index for Q2: {:.2f}".format(Q2_ratio))
print("Ratio-to-seasonality index for Q3: {:.2f}".format(Q3_ratio))
print("Ratio-to-seasonality index for Q4: {:.2f}".format(Q4_ratio))

```
